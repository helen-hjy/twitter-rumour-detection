```python
tab = '\t'

with open('../project-data/dev.data.txt', mode='r') as in_file, \
    open('../project-data/dev.label.txt', mode='r') as label_file, \
    open('../project-data/dev.data.tsv', mode='w') as out_file:

    labels = label_file.readlines()

    out_file.write('source\treplies\tlabel\n')

    for i, line in enumerate(in_file.readlines()):
        line = line.split(",")
        for j, tweet_id in enumerate(line):
            tweet = client.get_tweet(tweet_id.strip())
            if tweet.data:
                text = tweet.data.text.replace('\n', ' ').replace('\r', '')
                out_file.write(f'{text}{tab if j == 0 else "<EOS>"}')
            else:
                # if source is not exist
                if j == 0:
                    break
        
        print(i+1, "done")
        out_file.write(f'{tab + labels[i].strip()}\n')
        out_file.flush()    
```
