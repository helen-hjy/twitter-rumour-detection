```python
import json
def get_tweet(tweet_id):
    path = "/content/tweet-objects/" + tweet_id + ".json"
    with open(path,'r') as f:
        load_dict = json.load(f)
        return load_dict
```

```python
from os import cpu_count
tab = '\t'
with open('/content/test.data.txt') as f,\
    open('/content/test.data.tsv', mode='w') as out_file:
    out_file.write('source\treplies\tlabel\n')
    count = 0
    for line in f:
        count += 1
        line = line.split(",")

        for j, tweet_id in enumerate(line):

            tweet = get_tweet(tweet_id.strip())

            text = tweet['text']
            text = text.replace('\n', ' ').replace('\r', '').replace(tab, ' ').replace('"', '')


            out_file.write(f'{text}{ tab if j == 0 else "<EOS>"}')

        out_file.write(f'{tab}0\n')
        out_file.flush()

    print("total tweets: " + str(count))
```
